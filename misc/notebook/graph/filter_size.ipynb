```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
})
```

```python
A = [0.6194, 0.6821, 0.6404, 0.4643, 0.2064]
E = [0.2371, 0.3075, 0.3109, 0.2571, 0.2059]
size = [256, 64, 24, 8, 0]
```

```python

plt.plot(size, A)
plt.fill_between(size, E, A, alpha=0.25, color='C0', label=r'$\kappa$ Aptos')

plt.scatter(size, A, color='C0', s=10)

plt.plot(size, E)
plt.fill_between(size, [0]*len(size), E, alpha=0.25, color='C1', label=r'$\kappa$ EyePacs')

plt.scatter(size, E, color='C1', s=10)

plt.legend(loc='upper right')
plt.ylim(0, 0.8)
plt.xlim(0, 256)
plt.grid(alpha=0.3)
plt.ylabel(r'$\kappa$')
fig = plt.gcf().set_size_inches((10, 6))

plt.xlabel('Taille de lésions minimale (en pixels)')
plt.savefig('filter_size_effect_idrid.png')
```
