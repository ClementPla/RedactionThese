```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
import bokeh
import datetime
```

```python
n_papers = [3000, 4030,2810,2170,1590,1140,869,740,679, 616,545,476][::-1]
start = 2012
x = np.arange(2012, start+len(n_papers))

xlabels = [f'{y}-{y+1}' for y in x]
ax = plt.subplot()
ax.bar(x, n_papers, alpha=0.5, edgecolor='black')
ax.plot(x, n_papers, '--', color='black', linewidth=1)

ax.set_xticks(x, labels=xlabels, )
plt.xticks(rotation=25)
plt.xlabel('Ann√©e')
plt.ylabel("Nombre d'articles de revue")

plt.savefig('paper_published.png')
```
